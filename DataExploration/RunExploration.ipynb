```python
from RunPlotting import *
import os
from random import sample
```

```python
# Support Of Each Class
INITALDATAPATH = '../SortedData/'

LabelFolders = os.listdir(INITALDATAPATH)

for Folder in LabelFolders:
    FullPath = os.path.join(INITALDATAPATH, Folder)
    
    AllFiles = os.listdir(FullPath)
    FileCount = len(AllFiles)
    
    print(f"Label Class: {Folder}\nLabel Count: {FileCount}")
```

```python
for Folder in LabelFolders:
    FullPath = os.path.join(INITALDATAPATH, Folder)
    
    FileList = os.listdir(FullPath)
    
    # Create List For Sampling
    NumberList = list(range(0, len(FileList) - 1))
    
    # Random Sample
    RandomSample = sample(NumberList, 5)
    
    # Plot Each Sample
    for Sample in RandomSample:
        PlottingFilePath = os.path.join(FullPath, FileList[Sample])
        plot_file(PlottingFilePath)
```
